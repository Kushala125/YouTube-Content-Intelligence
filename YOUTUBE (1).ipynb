```python
import pandas as pd
import numpy as np

# 1. Load the data FIRST
df = pd.read_csv("US_Trending.csv") 

# 2. Clean the titles (removes those "..." rows)
df = df[df['title'].str.contains(r'[a-zA-Z]', na=False)]

# 3. CREATE the comment_count column (Synthetic)
# We make this first so the next steps don't crash
np.random.seed(42)
df['comment_count'] = (df['views'] * np.random.uniform(0.001, 0.005, len(df))).astype(int)

# 4. NOW calculate the engagement rate (because comment_count now exists!)
df['engagement_rate'] = ((df['likes'] + df['comment_count']) / df['views']) * 100

# 5. Fix the dates and round
df['publish_time'] = pd.to_datetime(df['publish_time'])
df['engagement_rate'] = df['engagement_rate'].round(2)

print(df[['title', 'views', 'likes', 'comment_count', 'engagement_rate']].head())
```

```python
df
```

```python
# Fix date
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')

# Fix missing description
df['description'] = df['description'].fillna('No description')

# Fix duplicate columns
df.drop(columns=['comment_count'], inplace=True)
df.rename(columns={'comments': 'comment_count'}, inplace=True)

# Fix engagement rate
df['engagement_rate'] = (df['likes'] + df['comment_count']) / df['views']
```

```python
#CLEAN description
df['description'] = df['description'].fillna('No description')
df['desc_length'] = df['description'].apply(len)
df['has_description'] = df['description'].apply(lambda x: 0 if x == 'No description' else 1)
```

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df['category_name'], df['high_engagement'])

chi2, p, dof, expected = chi2_contingency(contingency)

print("P-value:", p)
```

```python
#What range likely contains the true average views?


#The confidence interval provides a range where the true mean is expected to lie with 95% certainty. This helps quantify uncertainty in the data.

import scipy.stats as stats

mean = df['views'].mean()
sem = stats.sem(df['views'])

ci = stats.t.interval(0.95, len(df)-1, loc=mean, scale=sem)

print(ci)
```

```python
from scipy.stats import ttest_ind
morning = df[df['publish_hour'] < 12]['views']
evening = df[df['publish_hour'] >= 12]['views']

t_stat, p_value = ttest_ind(morning, evening)

print("T-stat:", t_stat)
print("P-value:", p_value)
```

```python
high = df[df['high_engagement'] == True]['views']
low = df[df['high_engagement'] == False]['views']

from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(high, low)

print("T-stat:", t_stat)
print("P-value:", p_value)
print("High mean:", high.mean())
print("Low mean:", low.mean())
```

```python
from scipy.stats import ttest_ind

tagged = df[df['has_tags'] == 1]['views']
no_tags = df[df['has_tags'] == 0]['views']

t_stat, p_value = ttest_ind(tagged, no_tags)

print("T-stat:", t_stat)
print("P-value:", p_value)
print("Tagged mean:", tagged.mean())
print("No Tags mean:", no_tags.mean())
```

```python
from scipy.stats import ttest_ind

weekday = df[df['publish_day'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday'])]['views']
weekend = df[df['publish_day'].isin(['Saturday','Sunday'])]['views']

t_stat, p_value = ttest_ind(weekday, weekend)

print("T-stat:", t_stat)
print("P-value:", p_value)
print("Weekday mean:", weekday.mean())
print("Weekend mean:", weekend.mean())
```

```python
from scipy.stats import ttest_ind

desc = df[df['has_description'] == 1]['views']
no_desc = df[df['has_description'] == 0]['views']

t_stat, p_value = ttest_ind(desc, no_desc)

print("T-stat:", t_stat)
print("P-value:", p_value)
print("With description mean:", desc.mean())
print("No description mean:", no_desc.mean())
```

```python
import plotly.graph_objects as go

fig = go.Figure()

hours = sorted(df['publish_hour'].unique())

for i, hour in enumerate(hours):
    temp = df[df['publish_hour'] == hour]

    fig.add_trace(go.Funnel(
        visible=(i == 0),  # only first visible
        y=['Views', 'Likes', 'Comments'],
        x=[
            temp['views'].sum(),
            temp['likes'].sum(),
            temp['comment_count'].sum()
        ],
        textinfo="value+percent previous",
        name=f"Hour {hour}"
    ))

# 🎯 SLIDER CONTROL
steps = []
for i, hour in enumerate(hours):
    step = dict(
        method="update",
        args=[{"visible": [j == i for j in range(len(hours))]}],
        label=f"{hour}:00"
    )
    steps.append(step)

fig.update_layout(
    title="🎬 Moving Funnel — Engagement Flow Over Time",
    sliders=[dict(
        active=0,
        steps=steps,
        currentvalue={"prefix": "Hour: "}
    )]
)

fig.show()
fig.update_layout(
    updatemenus=[{
        "type": "buttons",
        "buttons": [
            {"label": "▶ Play", "method": "animate", "args": [None]}
        ]
    }]
)
```
